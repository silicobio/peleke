# Generate Training Dataset

```python
import pandas as pd
import os
```

```python
seq_df = pd.read_csv('sabdab_sequences.csv')

## Create 'antibody_sequences' column with '|' delimiter
seq_df['antibody_sequences'] = seq_df['h_chain_seq'] + '|' + seq_df['l_chain_seq']

seq_df
```

```python
## Loop through the CSVs in the './contacts' directory and combine them into a single DataFrame
contacts_directory = './contacts'

epi_df = pd.DataFrame()
for filename in os.listdir(contacts_directory):
    if filename.endswith('_contacts.csv'):
        file_path = os.path.join(contacts_directory, filename)
        row_df = pd.read_csv(file_path)
        epi_df = pd.concat([epi_df, row_df], ignore_index=True)

## Remove '.pdb' from 'pdb_id' column
epi_df['pdb_id'] = epi_df['pdb_id'].str.replace('.pdb', '')

## Write the DataFrame to a new CSV file
epi_df.to_csv('sabdab_highlighted_epitopes.csv', index=False)

epi_df
```

```python
## Join the two DataFrames on 'pdb_id' and 'antigen_ids'
joined_df = pd.merge(seq_df, epi_df, on=['pdb_id', 'antigen_ids'], how='left')

## Remove rows with NaN in 'antigen_ids' column
joined_df = joined_df[~joined_df['h_chain_seq'].isna()]
joined_df = joined_df[~joined_df['l_chain_seq'].isna()]
joined_df = joined_df[~joined_df['antigen_seqs'].isna()]

## Remove rows where 'highlighted_epitope_sequences' is '|'
joined_df = joined_df[joined_df['highlighted_epitope_sequences'] != '|']

## TEMP: Remove rows with NaN in 'highlighted_epitope_sequences' column
joined_df = joined_df[~joined_df['highlighted_epitope_sequences'].isna()]

joined_df
```

```python
## Write to CSV
joined_df.to_csv('sabdab_training_dataset.csv', index=False)
```
